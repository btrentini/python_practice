# Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="ticks", palette='deep')
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Univariate

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') # Combine two distribution plots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # Combine two distribution plots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') # Combine two distribution plots
```

# PAIR PLOT

```python
sns.pairplot(tips,height=5, hue='sex',palette='cubehelix',kind='scatter')
```

# RUG PLOT

```python
sns.rugplot(tips['total_bill'])
```

```python
# For each point above, there's a normal distribution courve - whenyou sum all normal distributions, you get KDE

# KDE line based on Rug Plot
# Kernel Density Estimation plots.....

# Graussian Distribution: normal distribution
```
