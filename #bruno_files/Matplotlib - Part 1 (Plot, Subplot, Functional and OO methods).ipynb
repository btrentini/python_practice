```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

# Functional Method

```python
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('TITLE HERE')
```

```python
# MULTIPLOT ON SAME CANVAS
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
# Number of rows, columns and level
```

# OBJECT ORIENTED METHOD

```python
# Create Figure objkects and then call methods
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1.8,0.8]) #[ Left, Bottom, Width, Heighy ]
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.75,0.15,0.7,0.45])

axes2.plot(y,x,'g-^')
axes1.plot(x,y,'r--')

axes2.set_title('Detail')
axes1.set_title('MAIN THING')
```
