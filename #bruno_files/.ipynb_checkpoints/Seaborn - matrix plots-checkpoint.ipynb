# Matrix Plots (Heatmaps, for example)

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

# Heatmap
#### Data must be in matrix form. 
<p> index name and column name match ups the cell value indicate something relevant for both of those names </p> 

```python
# Resolve: pivot or correlation
```

```python
tc = tips.corr() # Matrix Form
```

```python
sns.heatmap(tc, annot=True, cmap='cubehelix')
```

tc

```python
flights.head()
```

```python
# Flights: must convert one to be the index
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(fp, annot=True,fmt='.5g',linecolor='white',linewidths=3)
```

# Cluster Map

```python
sns.clustermap(fp, cmap='coolwarm',standard_scale=1)
```
