```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

# See distributions and reference them to numerical / categorical column

```python
import numpy as np
sns.barplot(x=tips['sex'], y=tips['total_bill'], data=tips, estimator=np.sum)
#Average (mean) by category
```

```python
sns.countplot(x='sex', data=tips) #Same as pandas.count
```

# BOXPLOT

```python
sns.boxplot(x='day', y='total_bill',data=tips, hue='smoker')
```

# Violin Plots

```python
sns.violinplot(x='day', y='total_bill',data=tips,hue='sex',split=True)
```

# Strip Plot

```python
sns.stripplot(x='day', y='total_bill',data=tips, jitter=True, hue='smoker', split=True)
```

# Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='pastel')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='pastel')
```

# Factor Plot

```python
sns.factorplot(x='day', y='total_bill', data=tips,kind='bar')
```
