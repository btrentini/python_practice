# Scatter

```python
def convert_positive(x):
    if x < 0:
        return x*-1
    else:
        return x
    
df1['X'] = df1['C'].apply(convert_positive)
```

```python
df1['Y'] = df1['C'].apply(lambda x: x*-1 if x<0 else x)
```

```python
df1.head()
```

```python
df1.plot.scatter(x='A', y='B',c='C', s=df1['X']*150, figsize=(18,10))
```
