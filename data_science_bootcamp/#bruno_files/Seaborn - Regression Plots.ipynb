# Regression Plots

```python
#LMPLOT - Linear Models with Seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['o','v'],
          scatter_kws={'s':100},height=8)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='smoker',hue='sex',aspect=0.6)
```
