```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#Tuple unpacking


axes[0].plot(x,y)
axes[0].set_title('See it')

axes[1].plot(y,x)
axes[1].set_title('See it again')
```

```python
type(axes)
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,4), dpi=100) #DPI = Pixels per Inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4), dpi=100) #DPI = Pixels per Inch

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Saving Figure

```python
fig.savefig('my_plot.png',dpi=200)
```

# Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='Square', lw=8, alpha=0.1)
ax.plot(x,x**3, label='Cubic')

#ax.legend(loc=3)
ax.legend(loc=(0.1,0.1))
```
