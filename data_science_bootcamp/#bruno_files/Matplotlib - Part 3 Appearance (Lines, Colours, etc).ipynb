# Plot appearance

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#99cc00',lw=8, alpha=0.95, ls='--', marker='s', markersize=12,
        markerfacecolor='#990000', markeredgewidth=3,markeredgecolor='#000000')
```

# Set xlim

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#99cc00',lw=8, alpha=0.95, ls='--', marker='s', markersize=12,
        markerfacecolor='#990000', markeredgewidth=3,markeredgecolor='#000000')

ax.set_xlim([0,1])
ax.set_ylim([0,1.25])
```
