### The K Means Algorithms
<ul>
    <li> Choose a number of clusters K</li>
    <li> Randomly assign each point to a cluster </li>
    <li> Until clusters stop changing, repeat: </li>
    <ul> For each cluster, computer the cluster centroid by taking the mean vector of points in the cluster </ul>
    <ul> Assign each data point to the cluster for which the centrois is the closest </ul>

# Choosing K Value

```python
# No simple answer. No easy way.
# Elbow method: compute the sum of square error (SSE) for some vlaues of K (eg: 2,4,6,8)
# The SSE s defined as the sum of the squared distance between each member of the cluster and its centrois

# If you plot K against SSE, error decreases as k gets larger. This is becayse wgeb the number
# Of clusters increases, they should be smaller, so distortion is also smaller

# The idea of the elbow method is to choose the K at which the SSE decreases abruptuly
# This produces an "elbow effect" in the graph
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers = 4, cluster_std=1.8,random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Set1' )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# Real data.... it's done at this stage. Cuz we don't have labels
# K means is unsupervised, so you can't classify, but find labels
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,sharey=True, figsize=(13,5))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='Set1')

ax2.set_title('Orginal')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Set1')
```
